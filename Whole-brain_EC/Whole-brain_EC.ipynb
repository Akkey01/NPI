```python
import numpy as np
import matplotlib.pyplot as plt
from nilearn.plotting import plot_matrix
from matplotlib.colors import TwoSlopeNorm
```

```python
plt.rcParams['figure.dpi'] = 200
plt.rcParams['font.size'] = 8.0
```

### MNI-MSDL
The element EC\[i][j] represents the effective connectivity from brain region i to brain region j.

```python
EC = np.loadtxt('./MNI-MSDL/EC_MNI-MSDL.txt')
with open('./MNI-MSDL/labels_MNI-MSDL.txt') as f: labels = f.read().splitlines()
```

```python
divnorm = TwoSlopeNorm(vmin = -0.1, vcenter = 0, vmax = 0.2)
fig = plot_matrix(EC, norm = divnorm, labels = labels)
plt.show()
```
