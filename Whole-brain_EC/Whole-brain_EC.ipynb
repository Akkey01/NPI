```python
import numpy as np
import matplotlib.pyplot as plt
from nilearn.plotting import plot_matrix
from matplotlib.colors import TwoSlopeNorm
```

```python
plt.rcParams['figure.dpi'] = 200
plt.rcParams['font.size'] = 8.0
```

### MNI-MSDL
The element EC\[i][j] represents the effective connectivity from brain region i to brain region j.

```python
EC = np.loadtxt('./MNI-MSDL/EC_MNI-MSDL.txt')
with open('./MNI-MSDL/labels_MNI-MSDL.txt') as f: labels = f.read().splitlines()
```

```python
divnorm = TwoSlopeNorm(vmin = -0.5, vcenter = 0, vmax = 1.0)
fig = plot_matrix(EC, norm = divnorm, labels = labels, colorbar = False)
plt.show()
```

### MNI-MMP1
The element EC\[i][j] represents the effective connectivity from brain region i to brain region j.

```python
EC = np.loadtxt('./MNI-MMP1/EC_MNI-MMP1.txt')
with open('./MNI-MMP1/labels_MNI-MMP1.txt') as f: labels = f.read().splitlines()

neworderlr = np.loadtxt('./MNI-MMP1/neworderlr.txt')
neworder = np.array([np.where(neworderlr == i)[0][0] for i in range(len(neworderlr))])
oldorder = np.array([np.where(neworder == i)[0][0] for i in range(len(neworder))])
```

```python
divnorm = TwoSlopeNorm(vmin = -0.2, vcenter = 0, vmax = 1.0)
fig = plot_matrix(EC[neworder].T[neworder].T, norm = divnorm, colorbar = False)
plt.show()
```
